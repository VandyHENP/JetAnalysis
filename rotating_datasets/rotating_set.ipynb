```python
import pandas as pd
import numpy as np
```

```python
hadron_level_data = pd.read_csv("hadron_level_events.csv")
parton_level_data = pd.read_csv("parton_level_events.csv")
```

```python
hadron_level_data = hadron_level_data.drop(columns=["Unnamed: 91"]).dropna()
parton_level_data = parton_level_data.drop(columns=["Unnamed: 61"]).dropna()
```

```python
hadron_level_data
```

```python
x = hadron_level_data.values
y = parton_level_data.values
```

```python
x[1]
```

```python
def rotate(event, phi):   
    first = event[0]
    event = event[1:]
    index = 0
    for x in event:
        if (((index+1)%3 == 0) and (event[index] != 0)):
            event[index] = round((event[index] + phi + np.pi) % (2 * np.pi) - np.pi, 5) # doing this cuz csv file is mostly to 3 DP
        index+=1
    event = first + event 
    
```

```python
phi = np.pi/2
for hadron_event in x:
    rotate(hadron_event, phi)
for parton_event in y:
    rotate(parton_event, phi)
```

```python
hadron = pd.DataFrame(x)
parton = pd.DataFrame(y)
```

```python
hadron.to_csv("hadron_level_events_rotated_half_pi.csv")
parton.to_csv("parton_level_events_rotated_half_pi.csv")
```
